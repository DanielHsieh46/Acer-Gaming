```python
import yfinance as yf

ticker = yf.Ticker("6776.TW")

IS = ticker.quarterly_income_stmt
IS = IS.get("2024-06-30")

IS_2023 = ticker.income_stmt
IS_2023 = IS_2023.get("2023-12-31")

BS = ticker.quarterly_balance_sheet
BS1 = BS.get("2024-06-30")
BS2 = BS.get("2024-03-31")


Gross_Margin = "營業毛利率= " + str(IS.get("Gross Profit")/IS.get("Total Revenue")*100)
Operating_Margin = "營業利益率= " + str((IS.get("Operating Income")-IS.get("Write Off")) / IS.get("Total Revenue") * 100)
Net_Profit_Margin = "淨利率= " + str(IS.get("Net Income Continuous Operations") / IS.get("Total Revenue") * 100)
EPS = "EPS= " + str(IS.get("Basic EPS"))
ROA = "ROA= " + str(IS.get("Net Income Continuous Operations") / ((BS1.get("Total Assets")+BS2.get("Total Assets"))/2) * 100)
ROE = "ROE= " + str(IS.get("Net Income Continuous Operations") / ((BS1.get("Total Equity Gross Minority Interest") +BS2.get("Total Equity Gross Minority Interest"))/2) * 100)
Asset_Turnover = "資產周轉率= " + str(IS.get("Total Revenue") / ((BS1.get("Total Assets")+BS2.get("Total Assets"))/2))
Inventory_Turnover = "存貨周轉率= " + str(IS.get("Cost Of Revenue") / ((BS1.get("Inventory")+BS2.get("Inventory"))/2))
AR_Turnover = "*應收帳款週轉率= " + str(IS.get("Total Revenue") / ((BS1.get("Accounts Receivable")+BS1.get("Other Receivables")+BS2.get("Accounts Receivable")+BS2.get("Other Receivables"))/2))
Assets = "資產= " + str(BS1.get("Total Assets"))
Liabilities_Percentage = "負債%= " + str(BS1.get("Total Liabilities Net Minority Interest") / BS1.get("Total Assets") * 100)
Equity_Percentage = "業主權益%="  + str(BS1.get("Total Equity Gross Minority Interest") / BS1.get("Total Assets") * 100)

dividends = ticker.dividends
dividends_2024 = "2024發放股利= " + str(dividends[dividends.index.year == 2024].iloc[0])
EPS_2023 = "2023eps= " + str(IS_2023.get("Basic EPS"))

print(Gross_Margin)
print(Operating_Margin)
print(Net_Profit_Margin)
print(EPS)
print(ROA)
print(ROE)
print(Asset_Turnover)
print(Inventory_Turnover)
print(AR_Turnover)
print(Assets)
print(Liabilities_Percentage)
print(Equity_Percentage)
print(dividends_2024)
print(EPS_2023)
```
